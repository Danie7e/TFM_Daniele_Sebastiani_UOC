```python
import os
import pandas as pd

ruta_resultados_oligosacaridos = "/content/drive/MyDrive/TFM/T1/df_resultados_oligosacaridos_ramificados"
ruta_df_harm = "/content/drive/MyDrive/TFM/T1"

df_harm_file = os.path.join(ruta_df_harm,"df_harm_filtro_3.csv")
df_harm = pd.read_csv(df_harm_file)
df_harm.head()
```

```python
#@title Unir resultados oligoscaridos ramificados

output_folder = '/content/drive/MyDrive/TFM/T1'
# Crear lista para almacenar los DataFrames
lista_df = []

# Verificar existencia de la carpeta de input
if os.path.exists(ruta_resultados_oligosacaridos):
    # Escanear los archivos en la carpeta
    for i in range(1,10):
      file_path = os.path.join(ruta_resultados_oligosacaridos, f"oligosacaridos_ramificados_{i}.csv")

      # Leer el archivo y agregarlo a la lista
      df = pd.read_csv(file_path, sep=',')
      lista_df.append(df)  # Agregar el DataFrame a la lista

# Unir todos los DataFrames en uno solo
if lista_df:  # Verificar que haya archivos CSV procesados
    df_final = pd.concat(lista_df, ignore_index=True)

    # Definir la ruta del archivo de salida
    output_file = os.path.join(output_folder, "oligosacaridos_ramificados_tot.csv")

    # Guardar el DataFrame en un CSV
    df_final.to_csv(output_file, index=False)
    print(f"Archivo guardado en: {output_file}")
else:
    print("No se encontraron archivos CSV en la carpeta.")
```

```python
import os
import pandas as pd

output_folder = '/content/drive/MyDrive/TFM/T1'
ruta_resultados_oligosacaridos = '/content/drive/MyDrive/TFM/T1/df_resultados_oligosacaridos_ramificados'


lista_df = []


if os.path.exists(ruta_resultados_oligosacaridos):

    for archivo in os.listdir(ruta_resultados_oligosacaridos):
        if archivo.endswith('.csv'):
            file_path = os.path.join(ruta_resultados_oligosacaridos, archivo)
            df = pd.read_csv(file_path, sep=',')
            lista_df.append(df)


if lista_df:
    df_final = pd.concat(lista_df, ignore_index=True)


    output_file = os.path.join(output_folder, "oligosacaridos_ramificados_tot.csv")


    df_final.to_csv(output_file, index=False)
    print(f"Archivo guardado en: {output_file}")
else:
    print("No se encontraron archivos CSV en la carpeta.")
```

```python
#@title Crear lista entrada y descartar

lista_oligosacaridos = df_final["PDB_entry_id"].tolist()
print(f"Elementos en df antes de aplicar el filtro: {df_harm.shape[0]}")

df_harm_filtro_4 = df_harm[~df_harm["PDB_entry_id"].isin(lista_oligosacaridos)]
print(f"Elementos en df después de aplicar el filtro: {df_harm_filtro_4.shape[0]}")
print(f"Se han descartado {len(lista_oligosacaridos)} entradas")
```

```python
#@title Guardar el archivo filtrado (Sin oligosacáridos ramificados)

output_folder = '/content/drive/MyDrive/TFM/T1'
output_file = os.path.join(output_folder, "df_harm_filtro_4.csv")
df_harm_filtro_4.to_csv(output_file, index=False)
print(f"Archivo guardado en: {output_file}")
```
