- Eliminar entradas que no tengan "Uniprot_id".

```python
#@title Abrir el dataframe

import os
import pandas as pd

input_folder = "/content/drive/MyDrive/TMF/T1/LIGANDO_ENTERRADO" # Ruta del df armonizado sin entradas covalentes
input_file = os.path.join(input_folder, "df_sin_covalentes.csv") # Nombre df armonizado sin entradas covalentes
df_harm = pd.read_csv(input_file, sep = ',')
print(df_harm.shape[0],df_harm.columns)
```

```python
#@title Quitar ligandos

ligandos_a_eliminar = ["MAN", "GAL", "BGC", "BMA", "NAG", "A2G"]

print(f"Dataframe antes del filtro: {df_harm.shape[0]}")

df_harm_filtro_1 = df_harm[~df_harm['Ligand_id'].isin(ligandos_a_eliminar)]

print(f"Dataframe despues del filtro: {df_harm_filtro_1.shape[0]}")
```

```python
df_harm_filtro_1['Uniprot_id'].count()
```

```python
#@title Descartar entradas sin Uniprot_id

print(f"Dataframe antes del filtro: {df_harm_filtro_1.shape[0]}")
df_harm_filtro_2 = df_harm_filtro_1[df_harm_filtro_1['Uniprot_id'].notna()]
print(f"Dataframe despues del filtro: {df_harm_filtro_2.shape[0]}")
```

```python
df_harm_filtro_2
```

## Ligandos Sin InChI


```python
lista_no_inchi = df_harm_filtro_2['Ligand_InChi'].isna()
df_sininchi = df_harm_filtro_2[lista_no_inchi]
df_sininchi
```

```python
df_sininchi['Ligand_id'].value_counts()
```

```python
df_unknown = df_harm_filtro_2[df_harm_filtro_2['Ligand_id']=='UNL']
df_unknown.shape[0]
```

Los ligandos sin código InChI corresponden a ligandos Unknown(UNL). Voy a descartarlos porque no van a ser útil para el análisis.

```python
#@title Descartar ligandos UNL (Unknown)

print(f"Dataframe antes del filtro: {df_harm_filtro_2.shape[0]}")
df_harm_filtro_3 = df_harm_filtro_2[df_harm_filtro_2['Ligand_id'] != 'UNL']
print(f"Dataframe despues del filtro: {df_harm_filtro_3.shape[0]}")
```
