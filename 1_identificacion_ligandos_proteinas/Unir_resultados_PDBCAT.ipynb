```python
#@title Crear tabla única con información entradas pdb
import os
import pandas as pd

# Rutas de las carpetas de input y output
input_folder = '/content/drive/MyDrive/TMF/T1/ENLACE_COVALENTE/RESULTADOS_PDB-CAT'
output_folder = '/content/drive/MyDrive/TMF/T1/ENLACE_COVALENTE'

# Crear lista para almacenar los DataFrames
lista_df = []

# Verificar existencia de la carpeta de input
if os.path.exists(input_folder):
    # Escanear los archivos en la carpeta
    for i in range(1,10):
      file_path = os.path.join(input_folder, f"df_ligands_{i}.csv")

      # Leer el archivo y agregarlo a la lista
      df = pd.read_csv(file_path, sep=',')
      lista_df.append(df)  # Agregar el DataFrame a la lista

# Unir todos los DataFrames en uno solo
if lista_df:  # Verificar que haya archivos CSV procesados
    df_final = pd.concat(lista_df, ignore_index=True)

    # Definir la ruta del archivo de salida
    output_file = os.path.join(output_folder, "df_entradas_cov_nocov.csv")

    # Guardar el DataFrame en un CSV
    df_final.to_csv(output_file, index=False)
    print(f"Archivo guardado en: {output_file}")
else:
    print("No se encontraron archivos CSV en la carpeta.")
```

```python
df_harm =df_harm.reset_index()
df_cov = df_cov.reset_index()
```

```python
#@title Lista entradas con enlaces covalentes

lista_covalent_entradas = []
lista_covalent_lig = []
for index_df_harm,pdb_dh_harm in enumerate(df_harm['PDB_entry_id']):
  for index_cov,pdb_cov in enumerate(df_cov['ID']):
    if pdb_dh_harm == pdb_cov.upper():
      lista_covalent_entradas.append(pdb_dh_harm)
      print(f"{pdb_dh_harm}, ligand: {df_cov.loc[index_cov,'Molecule']}")


      if df_harm.loc[index_df_harm,'Ligand_id'] == df_cov.loc[index_cov,'Molecule']:
          lista_covalent_lig.append(pdb_dh_harm)

print(len(lista_covalent_lig))
print(len(lista_covalent_entradas))
```

```python
print(len(lista_covalent_lig))
print(len(lista_covalent_entradas))
```

```python
print(f"Se van a descartar: {len(lista_covalent_lig)} entradas con enlaces covalentes")
```

## Ejemplo de que el código ha funcionado:

- En la lista "lista_covalent_entradas", hay entradas pdb que comparten el id pero que tienen ligandos diferentes. Esto significa que el enlace covalente se establece con un residuo diferente del ligando

- En la lista "lista_covalent_lig", hay entradas pdb que comparten el id y que tienen los mismos ligandos. Esto significa que el enlace covalente se establece con el ligando

```python
#@title Descartar entradas con enlaces convalentes

print(f"Entradas en df inicial: {df_harm.shape[0]}")
df_sin_covalentes = df_harm[~df_harm['PDB_entry_id'].isin(lista_covalent_lig)]
print(f"Entradas después del filtro: {df_sin_covalentes.shape[0]}")
```
