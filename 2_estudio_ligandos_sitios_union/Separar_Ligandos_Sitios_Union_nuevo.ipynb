```python
#@title Función para separar ligando y sitio de unión y guardarlos en archivos MMCIF distintos

def separar_ligando_sitio(row):
    from Bio.PDB import MMCIFParser, PDBIO, Select, is_aa
    import os
    import numpy as np

    try:
        parser = MMCIFParser(QUIET=True)
        input_path = "/content/drive/MyDrive/TFM/T2/CIF_DF"  # Archivos mmCIF de entrada
        output_path_pdb = "/content/drive/MyDrive/TFM/T2/TOT_ligandos_sitios/TOTAL/TOTAL_PDB/sitios_pdb_proteina"

        os.makedirs(output_path_pdb, exist_ok=True)

        id_pdb = row['PDB_entry_id']
        id_valor = id_pdb.lower()
        ligando = row['Ligand_id']
        coord = row['Coordenadas']
        cadena_lig = coord[2]
        id_res_lig = coord[3]

        archivo = os.path.join(input_path, f"{id_valor}.cif")  # Entrada .cif
        estructura = parser.get_structure(f"prot_{id_pdb}", archivo)

        class SelectLigand(Select):
            def __init__(self, ligand_residue):
                self.ligand_residue = ligand_residue

            def accept_residue(self, residue):
                return residue == self.ligand_residue

        class SelectSitioUnion(Select):
            def __init__(self, sitio_activo_residues):
                self.sitio_activo_residues = sitio_activo_residues

            def accept_residue(self, residue):
                return residue in self.sitio_activo_residues

        ligand_atoms = []
        ligand_res = None

        for modelo in estructura:
            for cadena in modelo:
                if cadena.id == cadena_lig:
                    for residuo in cadena:
                        if residuo.resname == ligando and residuo.id[1] == id_res_lig:
                            ligand_atoms.extend(list(residuo.get_atoms()))
                            ligand_res = residuo

        if ligand_res:
            io = PDBIO()
            io.set_structure(estructura)
            output_file = os.path.join(output_path_pdb, f"{id_pdb}_{ligando}_{cadena_lig}_{id_res_lig}_ligando.pdb")
            io.save(output_file, SelectLigand(ligand_res))
        else:
            print(f"Ligando no encontrado en {id_pdb} cadena {cadena_lig}, residuo {id_res_lig}")
            return False

        sitio_activo = []

        for modelo in estructura:
            for cadena in modelo:
                for residuo in cadena:
                    if residuo == ligand_res or not is_aa(residuo):
                        continue
                    for atom in residuo:
                        for lig_atom in ligand_atoms:
                            distancia = np.linalg.norm(atom.get_coord() - lig_atom.get_coord())
                            if distancia <= 4:
                                sitio_activo.append(residuo)
                                break
                        else:
                            continue
                        break

        if sitio_activo:
            io_prot = PDBIO()
            io_prot.set_structure(estructura)
            output_file_prot = os.path.join(output_path_pdb, f"{id_pdb}_{ligando}_{cadena_lig}_{id_res_lig}_sitio.pdb")
            io_prot.save(output_file_prot, SelectSitioUnion(sitio_activo))

        return True

    except Exception as e:
        print(f"Error en {row['PDB_entry_id']}: {e}")
        return False
```
